```python
from tabulate import tabulate

with open('validation-status-train.txt', 'w') as log_file:
    log_file.write('================= Train Data Validation =================\n\n')
    result_validate = []
    
    log_file.write("================= Check Data Shape=================\n\n")
    n_column = train_data.shape[1]
    n_row = train_data.shape[0]
    log_file.write(f'Data Shape for Train Data: {n_row} rows and {n_column} columns\n\n')

    get_cols = train_data.columns
    for col in get_cols:
        col_type = train_data[col].dtype
        sum_null = round(train_data[col].isnull().sum() * 100 / len(train_data))
        sum_duplicate = round(train_data.duplicated(keep=False).sum())

        result_validate.append([col, col_type, sum_null, sum_duplicate])
    
    headers_validate = ['Column Name', 'Data Type', 'Missing Value (%)', 'Duplicate Value (Count)']
    log_file.write(tabulate(result_validate, headers_validate, tablefmt='grid'))
    log_file.write("\n\n")
```
